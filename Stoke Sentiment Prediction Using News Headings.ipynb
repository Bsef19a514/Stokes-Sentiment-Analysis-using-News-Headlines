```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Stock-Sentiment-Analysis/master/Data.csv',encoding = 'unicode_escape')
df
```

```python
#Remember 0 in "label" means stokes prices  will fall and 1 means stokes will goes up
```

```python
df.describe()
```

# Handling missing values

```python
df=df.dropna(axis=0, how='any')
df.describe()
```

# Convertinf datatype of "Label" feature

```python
df['Label']=df['Label'].astype(int)
df.Label.dtypes
```

# Renaming column names

```python
newcolumns=['Date','Label']
for i in range(0,25):
    newcolumns.append(str(i))
newcolumns
```

```python
df.columns=newcolumns
```

```python
df.head(1)
```

# Converting each News Headine into lower case

```python
df1=df.copy()
for column in newcolumns[2:]:
    df1[column]=df1[column].str.lower()
df1.head(3)
```

# Spliting data into train and test

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

# Removing all sepecial characters from datasets

```python
train.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
test.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
train.head(2)
```

```python
X_train=train.drop(['Date','Label'],axis='columns')
X_test=test.drop(['Date','Label'],axis='columns')
```

```python
X_test.head(1)
```

```python
y_train=train['Label']
y_test=test['Label']
```

# Joining all the news headings of one row in single cell so that we can apply CountVectorizer classifier()

```python
train_headlines=[]
for i in range(len(X_train)):
    train_headlines.append(" ".join(str(x) for x in X_train.iloc[i,0:]))
train_headlines[1]
```

```python
test_headlines=[]
for i in range(len(X_test)):
    test_headlines.append(" ".join(str(x) for x in X_test.iloc[i,0:]))
test_headlines[1]
```

```python
len(train_headlines)
```

```python
len(test_headlines)
```

# Applying CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(ngram_range=(1,1)) #ngram_range=(2,2)
counvec_train_headlines=cv.fit_transform(train_headlines)
counvec_train_headlines
```

```python
counvec_test_headings=cv.transform(test_headlines)
counvec_test_headings
```

# Using SVM

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(counvec_train_headlines,y_train)
```

```python
svc.score(counvec_test_headings,y_test)
```

# Using Random Forest Clssifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
rfc.fit(counvec_train_headlines,y_train)
```

```python
rfc.score(counvec_test_headings,y_test)
```

```python
y_predict=rfc.predict(counvec_test_headings)
y_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
